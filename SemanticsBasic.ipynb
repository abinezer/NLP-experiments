```python
import spacy
```

```python
nlp = spacy.load('en_core_web_lg')
```

```python
nlp(u'lion').vector
```

```python
tokens = nlp(u'the lion belongs to the cat family. Cats are pets.')
```

```python
for token1 in tokens:
    for token2 in tokens:
        print(f'{token1.text:{10}} {token2.text:{10}} {token1.similarity(token2)}')
```
